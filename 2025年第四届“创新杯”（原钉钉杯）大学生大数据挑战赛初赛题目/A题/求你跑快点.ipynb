```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import lightgbm as lgb
from lightgbm import LGBMRegressor
import optuna
from optuna.integration import LightGBMPruningCallback
import shap
import joblib
from tqdm import tqdm
import os
import time

# ==================== å…¨å±€å­—ä½“è®¾ç½® (æç®€ç‰ˆ) ====================
# éµç…§æ‚¨çš„è¦æ±‚ï¼Œç§»é™¤äº†å¤æ‚çš„å­—ä½“è®¾ç½®å‡½æ•°ï¼Œé‡‡ç”¨æœ€ç®€æ´çš„æ–¹å¼é…ç½®å­—ä½“
# è®¾ç½®ä¸€ä¸ªæ”¯æŒä¸­æ–‡çš„é»˜è®¤å­—ä½“'SimHei'ï¼Œè¿™æ˜¯é¿å…ä¹±ç æœ€ç›´æ¥æœ‰æ•ˆçš„æ–¹æ³•
plt.rcParams['font.sans-serif'] = ['SimHei'] 
# è§£å†³åœ¨ä½¿ç”¨ä¸­æ–‡å­—ä½“æ—¶ï¼Œè´Ÿå·'-'æ˜¾ç¤ºä¸ºæ–¹å—çš„é—®é¢˜
plt.rcParams['axes.unicode_minus'] = False  


# ==================== ä¸»ç¨‹åºå…¥å£ ====================
if __name__ == "__main__":
    
    # --- å…¨å±€é…ç½® ---
    # å°†æ‰€æœ‰å¯é…ç½®å‚æ•°é›†ä¸­ç®¡ç†ï¼Œæ–¹ä¾¿ä¿®æ”¹å’Œç»´æŠ¤
    CONFIG = {
        'data_path': r'D:\æ¡Œé¢\2025å¹´ç¬¬å››å±Šâ€œåˆ›æ–°æ¯â€ï¼ˆåŸé’‰é’‰æ¯ï¼‰å¤§å­¦ç”Ÿå¤§æ•°æ®æŒ‘æˆ˜èµ›åˆèµ›é¢˜ç›®\2025å¹´ç¬¬å››å±Šâ€œåˆ›æ–°æ¯â€ï¼ˆåŸé’‰é’‰æ¯ï¼‰å¤§å­¦ç”Ÿå¤§æ•°æ®æŒ‘æˆ˜èµ›åˆèµ›é¢˜ç›®\Aé¢˜\data\train_data.csv',
        'report_path': './reports_rul_prediction', # æŠ¥å‘Šä¿å­˜çš„æ ¹ç›®å½•
        'random_state': 42,           # éšæœºç§å­ï¼Œç¡®ä¿ç»“æœå¯å¤ç°
        'reference_year': 2025,       # ç”¨äºè®¡ç®—è®¾å¤‡å¹´é¾„çš„å‚è€ƒå¹´ä»½
        'test_size': 0.2,             # æµ‹è¯•é›†åˆ’åˆ†æ¯”ä¾‹
        'target_col': 'Remaining_Useful_Life_days', # ç›®æ ‡å˜é‡åˆ—å
        'cols_to_drop': ['Machine_ID', 'Failure_Within_7_Days', 'Installation_Year'], # åˆå§‹è¦åˆ é™¤çš„åˆ—
        'n_trials': 50,               # Optunaè¶…å‚æ•°æœç´¢çš„è¯•éªŒæ¬¡æ•°
    }

    # --- æŠ¥å‘Šç›®å½•è®¾ç½® ---
    # åˆ›å»ºä¸€ä¸ªå¸¦æ—¶é—´æˆ³çš„ç‹¬ç«‹æ–‡ä»¶å¤¹æ¥ä¿å­˜æœ¬æ¬¡è¿è¡Œçš„æ‰€æœ‰äº§å‡ºç‰©
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    report_dir = os.path.join(CONFIG['report_path'], f"LGBM_run_{timestamp}")
    os.makedirs(report_dir, exist_ok=True)
    print(f"ğŸ“ æœ¬æ¬¡è¿è¡Œçš„æ‰€æœ‰æŠ¥å‘Šã€æ¨¡å‹å’Œå›¾è¡¨å°†ä¿å­˜åœ¨: {os.path.abspath(report_dir)}\n")

    # ==================== é˜¶æ®µ 1: æ•°æ®åŠ è½½ ====================
    print("="*50)
    print("[é˜¶æ®µ 1/8] æ­£åœ¨åŠ è½½æ•°æ®...")
    print("="*50)
    try:
        df = pd.read_csv(CONFIG['data_path'])
        print(f"âœ… æ•°æ®åŠ è½½æˆåŠŸï¼æ•°æ®é›†åŒ…å« {df.shape[0]} è¡Œ, {df.shape[1]} åˆ—ã€‚")
    except FileNotFoundError:
        print(f"âŒ é”™è¯¯ï¼šæ•°æ®æ–‡ä»¶æœªæ‰¾åˆ°ï¼è¯·æ£€æŸ¥è·¯å¾„æ˜¯å¦æ­£ç¡®: {CONFIG['data_path']}")
        exit()

    # ==================== é˜¶æ®µ 2: ç‰¹å¾å·¥ç¨‹ ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 2/8] æ­£åœ¨è¿›è¡Œç‰¹å¾å·¥ç¨‹...")
    print("="*50)
    df['Machine_Age'] = CONFIG['reference_year'] - df['Installation_Year']
    df['Power_Temp_Interaction'] = df['Power_Consumption_kW'] * df['Temperature_C']
    df['Maintenance_Frequency'] = df['Maintenance_History_Count'] / (df['Machine_Age'] + 1)
    print("   - å·²åˆ›å»ºæ–°ç‰¹å¾: 'Machine_Age', 'Power_Temp_Interaction', 'Maintenance_Frequency'")
    
    missing_ratio = df.isnull().mean()
    high_missing_cols = missing_ratio[missing_ratio > 0.8].index.tolist()
    if high_missing_cols:
        print(f"   - âš ï¸ å‘ç°ä»¥ä¸‹åˆ—çš„ç¼ºå¤±ç‡è¶…è¿‡80%ï¼Œå°†è‡ªåŠ¨åŠ å…¥åˆ é™¤åˆ—è¡¨: {high_missing_cols}")
        CONFIG['cols_to_drop'].extend(high_missing_cols)
        
    df.drop(columns=list(set(CONFIG['cols_to_drop'])), errors='ignore', inplace=True)
    print(f"   - å·²åˆ é™¤ä¸å†éœ€è¦çš„åˆ—ã€‚å‰©ä½™ç‰¹å¾æ•°: {df.shape[1] - 1}")
    print("âœ… ç‰¹å¾å·¥ç¨‹å®Œæˆã€‚")

    # ==================== é˜¶æ®µ 3: æ•°æ®åˆ’åˆ† ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 3/8] æ­£åœ¨åˆ’åˆ†è®­ç»ƒé›†ä¸æµ‹è¯•é›†...")
    print("="*50)
    y = df[CONFIG['target_col']]
    X = df.drop(columns=[CONFIG['target_col']])
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=CONFIG['test_size'], random_state=CONFIG['random_state']
    )
    print(f"âœ… æ•°æ®åˆ’åˆ†å®Œæˆã€‚")
    print(f"   - è®­ç»ƒé›†æ ·æœ¬æ•°: {X_train.shape[0]}")
    print(f"   - æµ‹è¯•é›†æ ·æœ¬æ•°: {X_test.shape[0]}")

    # ==================== é˜¶æ®µ 4: æ•°æ®é¢„å¤„ç† ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 4/8] æ­£åœ¨å¯¹æ•°æ®è¿›è¡Œé¢„å¤„ç†...")
    print("="*50)
    print("   [4.1] å¯¹æ•°å€¼ç‰¹å¾è¿›è¡Œç¼©å°¾å¤„ç†ä»¥å‡å°‘å¼‚å¸¸å€¼å½±å“...")
    numeric_cols = X_train.select_dtypes(include=np.number).columns
    bounds = {}
    for col in tqdm(numeric_cols, desc="     - è®¡ç®—å¹¶åº”ç”¨è¾¹ç•Œ"):
        q01, q99 = X_train[col].quantile([0.01, 0.99])
        bounds[col] = (q01, q99)
        X_train.loc[:, col] = np.clip(X_train[col], q01, q99)
        X_test.loc[:, col] = np.clip(X_test[col], q01, q99)

    print("\n   [4.2] å¯¹ç±»åˆ«ç‰¹å¾è¿›è¡Œç‹¬çƒ­ç¼–ç (One-Hot Encoding)...")
    categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
    if categorical_cols:
        X_train = pd.get_dummies(X_train, columns=categorical_cols, drop_first=True, dtype=float)
        X_test = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True, dtype=float)
        X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0.0)
        print(f"     - å®Œæˆå¯¹ {categorical_cols} çš„ç¼–ç ã€‚")
    else:
        print("     - æ•°æ®é›†ä¸­æ— ç±»åˆ«ç‰¹å¾ï¼Œè·³è¿‡æ­¤æ­¥éª¤ã€‚")
    
    feature_names = list(X_train.columns) 
    print("âœ… æ•°æ®é¢„å¤„ç†å®Œæˆã€‚")

    # ==================== é˜¶æ®µ 5: Optuna è¶…å‚æ•°è°ƒä¼˜ ====================
    print("\n" + "="*50)
    print(f"[é˜¶æ®µ 5/8] æ­£åœ¨ä½¿ç”¨Optunaè¿›è¡Œ {CONFIG['n_trials']} è½®è¶…å‚æ•°è°ƒä¼˜...")
    print("="*50)
    def objective(trial):
        X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=CONFIG['random_state'])
        params = {
            'objective': 'regression_l1', 
            'metric': 'rmse',
            'n_estimators': trial.suggest_int("n_estimators", 400, 2500),
            'learning_rate': trial.suggest_float("learning_rate", 1e-3, 0.1, log=True),
            'num_leaves': trial.suggest_int("num_leaves", 20, 300),
            'max_depth': trial.suggest_int("max_depth", 5, 12),
            'subsample': trial.suggest_float("subsample", 0.6, 1.0), 
            'colsample_bytree': trial.suggest_float("colsample_bytree", 0.6, 1.0),
            'min_child_samples': trial.suggest_int("min_child_samples", 5, 100),
            'reg_alpha': trial.suggest_float("reg_alpha", 1e-3, 10.0, log=True), 
            'reg_lambda': trial.suggest_float("reg_lambda", 1e-3, 10.0, log=True),
            'random_state': CONFIG['random_state'],
            'verbose': -1,
            'n_jobs': -1
        }
        model = LGBMRegressor(**params)
        pruning_cb = LightGBMPruningCallback(trial, "rmse")
        model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], eval_metric='rmse',
                  callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=False), pruning_cb])
        preds = model.predict(X_val)
        return np.sqrt(mean_squared_error(y_val, preds))

    study = optuna.create_study(direction="minimize", pruner=optuna.pruners.MedianPruner(n_warmup_steps=5))
    study.optimize(objective, n_trials=CONFIG['n_trials'], n_jobs=-1, show_progress_bar=True)
    
    best_params = study.best_params
    print("\nâœ¨ è°ƒä¼˜å®Œæˆï¼")
    print(f"   - å‘ç°çš„æœ€ä¼˜éªŒè¯é›†RMSE: {study.best_value:.4f}")
    print(f"   - æœ€ä½³å‚æ•°ç»„åˆ: {best_params}")
    print("âœ… è¶…å‚æ•°è°ƒä¼˜å®Œæˆã€‚")

    # ==================== é˜¶æ®µ 6: æ¨¡å‹è®­ç»ƒä¸ä¿å­˜ ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 6/8] æ­£åœ¨ä½¿ç”¨æœ€ä½³å‚æ•°è®­ç»ƒæœ€ç»ˆæ¨¡å‹...")
    print("="*50)
    final_model = LGBMRegressor(**best_params, random_state=CONFIG['random_state'], n_jobs=-1, verbose=-1)
    final_model.fit(X_train, y_train)
    model_path = os.path.join(report_dir, 'lgbm_rul_prediction_model.pkl')
    joblib.dump(final_model, model_path)
    print(f"âœ… æœ€ç»ˆæ¨¡å‹è®­ç»ƒå®Œæˆï¼Œå¹¶å·²ä¿å­˜è‡³: {model_path}")

    # ==================== é˜¶æ®µ 7: æ¨¡å‹è¯„ä¼° ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 7/8] æ­£åœ¨æµ‹è¯•é›†ä¸Šè¯„ä¼°æœ€ç»ˆæ¨¡å‹æ€§èƒ½...")
    print("="*50)
    predictions = final_model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, predictions)
    
    print("--- æœ€ç»ˆæ¨¡å‹åœ¨æµ‹è¯•é›†ä¸Šçš„æ€§èƒ½æŒ‡æ ‡ ---")
    print(f"   - ğŸ“Š å‡æ–¹è¯¯å·® (MSE):    {mse:.4f}")
    print(f"   - ğŸ¯ å‡æ–¹æ ¹è¯¯å·® (RMSE):  {rmse:.4f}")
    print(f"   - ğŸ“ˆ R-squared (RÂ²):    {r2:.4f}")
    print("âœ… æ¨¡å‹è¯„ä¼°å®Œæˆã€‚")

    # ==================== é˜¶æ®µ 8: ç»“æœå¯è§†åŒ– ====================
    print("\n" + "="*50)
    print("[é˜¶æ®µ 8/8] æ­£åœ¨ç”Ÿæˆå¹¶ä¿å­˜åˆ†æå›¾è¡¨...")
    print("="*50)

    # --- å›¾1: çœŸå®å€¼ vs. é¢„æµ‹å€¼ ---
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=y_test, y=predictions, alpha=0.6, edgecolor='k', label='é¢„æµ‹ç‚¹')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='ç†æƒ³æƒ…å†µ (y=x)')
    plt.xlabel('çœŸå®çš„å‰©ä½™ä½¿ç”¨å¯¿å‘½ (å¤©)', fontsize=12)
    plt.ylabel('é¢„æµ‹çš„å‰©ä½™ä½¿ç”¨å¯¿å‘½ (å¤©)', fontsize=12)
    plt.title('æ¨¡å‹é¢„æµ‹å€¼ vs. çœŸå®å€¼å¯¹æ¯”å›¾', fontsize=16, pad=20)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.savefig(os.path.join(report_dir, "figure_1_actual_vs_predicted.png"), dpi=300)
    plt.show()
    print("   - âœ… å·²ä¿å­˜ï¼šçœŸå®å€¼ vs. é¢„æµ‹å€¼å¯¹æ¯”å›¾")

    # --- å›¾2: ç‰¹å¾é‡è¦æ€§ (ä¼˜åŒ–ç‰ˆ) ---
    print("   - æ­£åœ¨ç”Ÿæˆä¼˜åŒ–ç‰ˆçš„ç‰¹å¾é‡è¦æ€§å›¾...")
    # ä»è®­ç»ƒå¥½çš„æ¨¡å‹ä¸­æå–ç‰¹å¾é‡è¦æ€§æ•°æ®
    importance_df = pd.DataFrame({
        'feature': final_model.feature_name_,
        'importance': final_model.feature_importances_
    }).sort_values('importance', ascending=False).head(20) # æ’åºå¹¶é€‰å–Top 20

    # ä½¿ç”¨ Seaborn æ‰‹åŠ¨ç»˜åˆ¶æ¡å½¢å›¾ï¼Œä»¥è·å¾—æ›´å¥½çš„æ§åˆ¶
    plt.figure(figsize=(12, 10)) # å¢å¤§ç”»å¸ƒå°ºå¯¸ä»¥å®¹çº³é•¿æ ‡ç­¾
    ax = sns.barplot(x='importance', y='feature', data=importance_df, palette='viridis_r')

    # è®¾ç½®æ ‡é¢˜å’Œæ ‡ç­¾
    ax.set_title('ç‰¹å¾é‡è¦æ€§ (åŸºäºå¢ç›Š, Top 20)', fontsize=18, pad=20)
    ax.set_xlabel('ç‰¹å¾é‡è¦æ€§ (å¢ç›Šå€¼)', fontsize=14)
    ax.set_ylabel('ç‰¹å¾åç§°', fontsize=14)

    # ä¼˜åŒ–åæ ‡è½´çš„æ˜¾ç¤ºæ ¼å¼
    ax.tick_params(axis='x', labelsize=12) # è°ƒæ•´xè½´åˆ»åº¦å­—ä½“å¤§å°
    ax.tick_params(axis='y', labelsize=12) # è°ƒæ•´yè½´åˆ»åº¦å­—ä½“å¤§å°
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ','))) # å°†xè½´æ•°å­—æ ¼å¼åŒ–ä¸ºåƒåˆ†ä½

    # ä½¿ç”¨ tight_layout ç¡®ä¿æ‰€æœ‰å…ƒç´ éƒ½èƒ½å®Œæ•´æ˜¾ç¤º
    plt.tight_layout()
    
    # ä¿å­˜å›¾è¡¨
    plt.savefig(os.path.join(report_dir, "figure_2_feature_importance_optimized.png"), dpi=300)
    plt.show()
    print("   - âœ… å·²ä¿å­˜ï¼šä¼˜åŒ–ç‰ˆç‰¹å¾é‡è¦æ€§å›¾")

    # --- å›¾3: SHAP æ‘˜è¦å›¾ ---
    print("   - æ­£åœ¨è®¡ç®—SHAPå€¼ï¼Œè¿™å¯èƒ½éœ€è¦ä¸€äº›æ—¶é—´...")
    explainer = shap.TreeExplainer(final_model)
    shap_values = explainer.shap_values(X_test)
    
    plt.figure()
    shap.summary_plot(shap_values, X_test, feature_names=feature_names, show=False, plot_size=None)
    fig = plt.gcf()
    fig.set_size_inches(12, 8)
    plt.title('SHAPæ‘˜è¦å›¾ï¼šç‰¹å¾å¯¹æ¨¡å‹è¾“å‡ºçš„å½±å“', fontsize=16, pad=20)
    plt.tight_layout()
    plt.savefig(os.path.join(report_dir, "figure_3_shap_summary_plot.png"), dpi=300)
    plt.show()
    print("   - âœ… å·²ä¿å­˜ï¼šSHAPæ‘˜è¦å›¾")

    print("\nğŸ‰ å…¨éƒ¨æµç¨‹æ‰§è¡Œå®Œæ¯•ï¼")
```
