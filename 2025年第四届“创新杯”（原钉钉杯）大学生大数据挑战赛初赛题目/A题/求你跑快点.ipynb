```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import lightgbm as lgb
from lightgbm import LGBMRegressor
import optuna
from optuna.integration import LightGBMPruningCallback
import shap
import joblib
from tqdm import tqdm
import os
import time

# ==================== 全局字体设置 (极简版) ====================
# 遵照您的要求，移除了复杂的字体设置函数，采用最简洁的方式配置字体
# 设置一个支持中文的默认字体'SimHei'，这是避免乱码最直接有效的方法
plt.rcParams['font.sans-serif'] = ['SimHei'] 
# 解决在使用中文字体时，负号'-'显示为方块的问题
plt.rcParams['axes.unicode_minus'] = False  


# ==================== 主程序入口 ====================
if __name__ == "__main__":
    
    # --- 全局配置 ---
    # 将所有可配置参数集中管理，方便修改和维护
    CONFIG = {
        'data_path': r'D:\桌面\2025年第四届“创新杯”（原钉钉杯）大学生大数据挑战赛初赛题目\2025年第四届“创新杯”（原钉钉杯）大学生大数据挑战赛初赛题目\A题\data\train_data.csv',
        'report_path': './reports_rul_prediction', # 报告保存的根目录
        'random_state': 42,           # 随机种子，确保结果可复现
        'reference_year': 2025,       # 用于计算设备年龄的参考年份
        'test_size': 0.2,             # 测试集划分比例
        'target_col': 'Remaining_Useful_Life_days', # 目标变量列名
        'cols_to_drop': ['Machine_ID', 'Failure_Within_7_Days', 'Installation_Year'], # 初始要删除的列
        'n_trials': 50,               # Optuna超参数搜索的试验次数
    }

    # --- 报告目录设置 ---
    # 创建一个带时间戳的独立文件夹来保存本次运行的所有产出物
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    report_dir = os.path.join(CONFIG['report_path'], f"LGBM_run_{timestamp}")
    os.makedirs(report_dir, exist_ok=True)
    print(f"📁 本次运行的所有报告、模型和图表将保存在: {os.path.abspath(report_dir)}\n")

    # ==================== 阶段 1: 数据加载 ====================
    print("="*50)
    print("[阶段 1/8] 正在加载数据...")
    print("="*50)
    try:
        df = pd.read_csv(CONFIG['data_path'])
        print(f"✅ 数据加载成功！数据集包含 {df.shape[0]} 行, {df.shape[1]} 列。")
    except FileNotFoundError:
        print(f"❌ 错误：数据文件未找到！请检查路径是否正确: {CONFIG['data_path']}")
        exit()

    # ==================== 阶段 2: 特征工程 ====================
    print("\n" + "="*50)
    print("[阶段 2/8] 正在进行特征工程...")
    print("="*50)
    df['Machine_Age'] = CONFIG['reference_year'] - df['Installation_Year']
    df['Power_Temp_Interaction'] = df['Power_Consumption_kW'] * df['Temperature_C']
    df['Maintenance_Frequency'] = df['Maintenance_History_Count'] / (df['Machine_Age'] + 1)
    print("   - 已创建新特征: 'Machine_Age', 'Power_Temp_Interaction', 'Maintenance_Frequency'")
    
    missing_ratio = df.isnull().mean()
    high_missing_cols = missing_ratio[missing_ratio > 0.8].index.tolist()
    if high_missing_cols:
        print(f"   - ⚠️ 发现以下列的缺失率超过80%，将自动加入删除列表: {high_missing_cols}")
        CONFIG['cols_to_drop'].extend(high_missing_cols)
        
    df.drop(columns=list(set(CONFIG['cols_to_drop'])), errors='ignore', inplace=True)
    print(f"   - 已删除不再需要的列。剩余特征数: {df.shape[1] - 1}")
    print("✅ 特征工程完成。")

    # ==================== 阶段 3: 数据划分 ====================
    print("\n" + "="*50)
    print("[阶段 3/8] 正在划分训练集与测试集...")
    print("="*50)
    y = df[CONFIG['target_col']]
    X = df.drop(columns=[CONFIG['target_col']])
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=CONFIG['test_size'], random_state=CONFIG['random_state']
    )
    print(f"✅ 数据划分完成。")
    print(f"   - 训练集样本数: {X_train.shape[0]}")
    print(f"   - 测试集样本数: {X_test.shape[0]}")

    # ==================== 阶段 4: 数据预处理 ====================
    print("\n" + "="*50)
    print("[阶段 4/8] 正在对数据进行预处理...")
    print("="*50)
    print("   [4.1] 对数值特征进行缩尾处理以减少异常值影响...")
    numeric_cols = X_train.select_dtypes(include=np.number).columns
    bounds = {}
    for col in tqdm(numeric_cols, desc="     - 计算并应用边界"):
        q01, q99 = X_train[col].quantile([0.01, 0.99])
        bounds[col] = (q01, q99)
        X_train.loc[:, col] = np.clip(X_train[col], q01, q99)
        X_test.loc[:, col] = np.clip(X_test[col], q01, q99)

    print("\n   [4.2] 对类别特征进行独热编码(One-Hot Encoding)...")
    categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
    if categorical_cols:
        X_train = pd.get_dummies(X_train, columns=categorical_cols, drop_first=True, dtype=float)
        X_test = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True, dtype=float)
        X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0.0)
        print(f"     - 完成对 {categorical_cols} 的编码。")
    else:
        print("     - 数据集中无类别特征，跳过此步骤。")
    
    feature_names = list(X_train.columns) 
    print("✅ 数据预处理完成。")

    # ==================== 阶段 5: Optuna 超参数调优 ====================
    print("\n" + "="*50)
    print(f"[阶段 5/8] 正在使用Optuna进行 {CONFIG['n_trials']} 轮超参数调优...")
    print("="*50)
    def objective(trial):
        X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=CONFIG['random_state'])
        params = {
            'objective': 'regression_l1', 
            'metric': 'rmse',
            'n_estimators': trial.suggest_int("n_estimators", 400, 2500),
            'learning_rate': trial.suggest_float("learning_rate", 1e-3, 0.1, log=True),
            'num_leaves': trial.suggest_int("num_leaves", 20, 300),
            'max_depth': trial.suggest_int("max_depth", 5, 12),
            'subsample': trial.suggest_float("subsample", 0.6, 1.0), 
            'colsample_bytree': trial.suggest_float("colsample_bytree", 0.6, 1.0),
            'min_child_samples': trial.suggest_int("min_child_samples", 5, 100),
            'reg_alpha': trial.suggest_float("reg_alpha", 1e-3, 10.0, log=True), 
            'reg_lambda': trial.suggest_float("reg_lambda", 1e-3, 10.0, log=True),
            'random_state': CONFIG['random_state'],
            'verbose': -1,
            'n_jobs': -1
        }
        model = LGBMRegressor(**params)
        pruning_cb = LightGBMPruningCallback(trial, "rmse")
        model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], eval_metric='rmse',
                  callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=False), pruning_cb])
        preds = model.predict(X_val)
        return np.sqrt(mean_squared_error(y_val, preds))

    study = optuna.create_study(direction="minimize", pruner=optuna.pruners.MedianPruner(n_warmup_steps=5))
    study.optimize(objective, n_trials=CONFIG['n_trials'], n_jobs=-1, show_progress_bar=True)
    
    best_params = study.best_params
    print("\n✨ 调优完成！")
    print(f"   - 发现的最优验证集RMSE: {study.best_value:.4f}")
    print(f"   - 最佳参数组合: {best_params}")
    print("✅ 超参数调优完成。")

    # ==================== 阶段 6: 模型训练与保存 ====================
    print("\n" + "="*50)
    print("[阶段 6/8] 正在使用最佳参数训练最终模型...")
    print("="*50)
    final_model = LGBMRegressor(**best_params, random_state=CONFIG['random_state'], n_jobs=-1, verbose=-1)
    final_model.fit(X_train, y_train)
    model_path = os.path.join(report_dir, 'lgbm_rul_prediction_model.pkl')
    joblib.dump(final_model, model_path)
    print(f"✅ 最终模型训练完成，并已保存至: {model_path}")

    # ==================== 阶段 7: 模型评估 ====================
    print("\n" + "="*50)
    print("[阶段 7/8] 正在测试集上评估最终模型性能...")
    print("="*50)
    predictions = final_model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, predictions)
    
    print("--- 最终模型在测试集上的性能指标 ---")
    print(f"   - 📊 均方误差 (MSE):    {mse:.4f}")
    print(f"   - 🎯 均方根误差 (RMSE):  {rmse:.4f}")
    print(f"   - 📈 R-squared (R²):    {r2:.4f}")
    print("✅ 模型评估完成。")

    # ==================== 阶段 8: 结果可视化 ====================
    print("\n" + "="*50)
    print("[阶段 8/8] 正在生成并保存分析图表...")
    print("="*50)

    # --- 图1: 真实值 vs. 预测值 ---
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=y_test, y=predictions, alpha=0.6, edgecolor='k', label='预测点')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='理想情况 (y=x)')
    plt.xlabel('真实的剩余使用寿命 (天)', fontsize=12)
    plt.ylabel('预测的剩余使用寿命 (天)', fontsize=12)
    plt.title('模型预测值 vs. 真实值对比图', fontsize=16, pad=20)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.savefig(os.path.join(report_dir, "figure_1_actual_vs_predicted.png"), dpi=300)
    plt.show()
    print("   - ✅ 已保存：真实值 vs. 预测值对比图")

    # --- 图2: 特征重要性 (优化版) ---
    print("   - 正在生成优化版的特征重要性图...")
    # 从训练好的模型中提取特征重要性数据
    importance_df = pd.DataFrame({
        'feature': final_model.feature_name_,
        'importance': final_model.feature_importances_
    }).sort_values('importance', ascending=False).head(20) # 排序并选取Top 20

    # 使用 Seaborn 手动绘制条形图，以获得更好的控制
    plt.figure(figsize=(12, 10)) # 增大画布尺寸以容纳长标签
    ax = sns.barplot(x='importance', y='feature', data=importance_df, palette='viridis_r')

    # 设置标题和标签
    ax.set_title('特征重要性 (基于增益, Top 20)', fontsize=18, pad=20)
    ax.set_xlabel('特征重要性 (增益值)', fontsize=14)
    ax.set_ylabel('特征名称', fontsize=14)

    # 优化坐标轴的显示格式
    ax.tick_params(axis='x', labelsize=12) # 调整x轴刻度字体大小
    ax.tick_params(axis='y', labelsize=12) # 调整y轴刻度字体大小
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ','))) # 将x轴数字格式化为千分位

    # 使用 tight_layout 确保所有元素都能完整显示
    plt.tight_layout()
    
    # 保存图表
    plt.savefig(os.path.join(report_dir, "figure_2_feature_importance_optimized.png"), dpi=300)
    plt.show()
    print("   - ✅ 已保存：优化版特征重要性图")

    # --- 图3: SHAP 摘要图 ---
    print("   - 正在计算SHAP值，这可能需要一些时间...")
    explainer = shap.TreeExplainer(final_model)
    shap_values = explainer.shap_values(X_test)
    
    plt.figure()
    shap.summary_plot(shap_values, X_test, feature_names=feature_names, show=False, plot_size=None)
    fig = plt.gcf()
    fig.set_size_inches(12, 8)
    plt.title('SHAP摘要图：特征对模型输出的影响', fontsize=16, pad=20)
    plt.tight_layout()
    plt.savefig(os.path.join(report_dir, "figure_3_shap_summary_plot.png"), dpi=300)
    plt.show()
    print("   - ✅ 已保存：SHAP摘要图")

    print("\n🎉 全部流程执行完毕！")
```
